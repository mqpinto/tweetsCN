```python
import tensorflow as tf

# convert to lower case
textos = ""
for i in range (0, data_gene['TextPost'].values.shape[0]):
    textos = textos + data_gene['TextPost'].values[i].lower()

words = []


for word in textos.split():
    if word != '.': # because we don't want to treat . as a word
        words.append(word)
words = set(words) # so that all duplicate words are removed
word2int = {}
int2word = {}
vocab_size = len(words) # gives the total number of unique words
for i,word in enumerate(words):
    word2int[word] = i
    int2word[i] = word

words
```

```python
# raw sentences is a list of sentences.
raw_sentences = textos.split('.')
sentences = []
for sentence in raw_sentences:
    sentences.append(sentence.split())
    
print(sentences[0]) 
#depois é necessário separar em tweets em vez de frases
```
